All operations below are realized in order to clean, uniformize, and unite the data inside a unique dataframe.

```python
# Join interventions1_cardiac, interventions2_cardiac, and interventions3_cardiac into a single dataframe
interventions_cardiac = pd.concat([interventions1_cardiac, interventions2_cardiac, interventions3_cardiac])

# Rename columns for consistency
cad9_cardiac.rename(columns={'Mission ID':'Mission ID', 'Service Name':'Service Name', 'Latitude permanence': 'Latitude Permanence', 'Longitude permanence': 'Longitude Permanence', 'Latitude intervention': 'Latitude Intervention', 'Longitude intervention': 'Longitude Intervention', 'Vector Type':'Vector Type'}, inplace=True)
interventions_cardiac.rename(columns={'Mission ID':'Mission ID', 'Service Name':'Service Name', 'Latitude permanence': 'Latitude Permanence', 'Longitude permanence': 'Longitude Permanence', 'Latitude intervention': 'Latitude Intervention', 'Longitude intervention': 'Longitude Intervention', 'Vector type':'Vector Type', 'Abandon reason':'Abandon Reason'}, inplace=True)
interventions_bxl2_cardiac.rename(columns={'Mission ID':'Mission ID', 'Service Name NL':'Service Name', 'Latitude permanence': 'Latitude Permanence', 'Longitude permanence': 'Longitude Permanence', 'Latitude intervention': 'Latitude Intervention', 'Longitude intervention': 'Longitude Intervention', 'Vector type NL':'Vector Type', 'Abandon reason NL':'Abandon Reason'}, inplace=True)
interventions_bxl_cardiac.rename(columns={'mission_id':'Mission ID', 'service_name':'Service Name', 'latitude_permanence': 'Latitude Permanence', 'longitude_permanence':'Longitude Permanence', 'latitude_intervention':'Latitude Intervention', 'longitude_intervention':'Longitude Intervention', 'vector_type':'Vector Type', 't0':'T0', 't1':'T1', 't2':'T2', 't3':'T3', 't4':'T4', 't5':'T5', 't6':'T6', 'abandon_reason':'Abandon Reason'}, inplace=True)
#print(interventions_cardiac)

# Convert date/time strings to pandas datetime objects (this operation is realized in several step, dependent on the time stamp, due to the non-uniformity of the data even within the different time stamps)
cad9_cardiac['T0'] = pd.to_datetime(cad9_cardiac['T0'], format='%Y-%m-%d %H:%M:%S.%f')
interventions_cardiac['T0'] = pd.to_datetime(interventions_cardiac['T0'], format='%d%b%y:%H:%M:%S')
interventions_bxl2_cardiac['T0'] = pd.to_datetime(interventions_bxl2_cardiac['T0'], format='%d%b%y:%H:%M:%S')
interventions_bxl_cardiac['T0'] = interventions_bxl_cardiac['T0'].astype(str).str[:19]
interventions_bxl_cardiac['T0'] = pd.to_datetime(interventions_bxl_cardiac['T0'], format='%Y-%m-%d %H:%M:%S')

cad9_cardiac['T1'] = pd.to_datetime(cad9_cardiac['T1'], format='%d%b%y:%H:%M:%S')
interventions_cardiac['T1'] = pd.to_datetime(interventions_cardiac['T1'], format='%d%b%y:%H:%M:%S')
interventions_bxl2_cardiac['T1'] = pd.to_datetime(interventions_bxl2_cardiac['T1'], format='%d%b%y:%H:%M:%S')
interventions_bxl_cardiac['T1'] = interventions_bxl_cardiac['T1'].astype(str).str[:19]
interventions_bxl_cardiac['T1'] = pd.to_datetime(interventions_bxl_cardiac['T1'], format='%Y-%m-%d %H:%M:%S', errors = 'coerce')

times_steps = ['T2', 'T3', 'T4', 'T5', 'T6']

for time_step in times_steps:
    cad9_cardiac[time_step] = cad9_cardiac[time_step].astype(str).str[:19]
    cad9_cardiac[time_step] = pd.to_datetime(cad9_cardiac[time_step], format='%Y-%m-%d %H:%M:%S', errors= 'coerce')
    interventions_cardiac[time_step] = interventions_cardiac[time_step].astype(str).str[:19]
    interventions_cardiac[time_step] = pd.to_datetime(interventions_cardiac[time_step], format='%Y-%m-%d %H:%M:%S', errors ='coerce')
    interventions_bxl2_cardiac[time_step] = pd.to_datetime(interventions_bxl2_cardiac[time_step], format='%d%b%y:%H:%M:%S')
    interventions_bxl_cardiac[time_step] = interventions_bxl_cardiac[time_step].astype(str).str[:19]
    interventions_bxl_cardiac[time_step] = pd.to_datetime(interventions_bxl_cardiac[time_step], format='%Y-%m-%d %H:%M:%S', errors = 'coerce')

# Add a new column 'Abandon Reason' to cad9_cardiac with value 'Overleden', and also overwrite the interventions_bxl2_cardiac with the same (maybe some issue with the filtering earlier??)
cad9_cardiac['Abandon Reason'] = 'Overleden'
interventions_bxl2_cardiac['Abandon Reason'] = 'Overleden'

# Add a new column to track the original file
cad9_cardiac['Original File'] = 'cad9_cardiac'
interventions_cardiac['Original File'] = 'interventions_cardiac'
interventions_bxl_cardiac['Original File'] = 'interventions_bxl_cardiac'
interventions_bxl2_cardiac['Original File'] = 'interventions_bxl2_cardiac'

# Concatenate all dataframes
frames = [
    cad9_cardiac[['Mission ID', 'Service Name', 'Latitude Permanence', 'Longitude Permanence', 'Latitude Intervention', 'Longitude Intervention', 'Vector Type', 'T0', 'T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'Abandon Reason', 'Original File']], 
    interventions_cardiac[['Mission ID', 'Service Name', 'Latitude Permanence', 'Longitude Permanence', 'Latitude Intervention', 'Longitude Intervention', 'Vector Type', 'T0', 'T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'Abandon Reason', 'Original File']], 
    interventions_bxl_cardiac[['Mission ID', 'Service Name', 'Latitude Permanence', 'Longitude Permanence', 'Latitude Intervention', 'Longitude Intervention', 'Vector Type', 'T0', 'T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'Abandon Reason', 'Original File']], 
    interventions_bxl2_cardiac[['Mission ID', 'Service Name', 'Latitude Permanence', 'Longitude Permanence', 'Latitude Intervention', 'Longitude Intervention', 'Vector Type', 'T0', 'T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'Abandon Reason', 'Original File']]
]
cardiac_interventions = pd.concat(frames)

# Replace values in the 'Vector Type' column
cardiac_interventions['Vector Type'] = cardiac_interventions['Vector Type'].replace({'Ambulance': 'AMB', 'Ambulance Event': 'AMB', 'MUG Event': 'MUG'})

## Correction of the latitude and longitude values
# Define the desired ranges
latitude_range = (48, 53)
longitude_range = (1, 8)

# Define a function to adjust latitude and longitude values
def adjust_coordinate(value, coordinate_range):
    if value < coordinate_range[0] or value > coordinate_range[1]:
        return value / 10
    return value

# Apply the adjustment function to Latitude Permanence and Latitude Intervention
while any((cardiac_interventions['Latitude Permanence'] < latitude_range[0]) | (cardiac_interventions['Latitude Permanence'] > latitude_range[1])):
    cardiac_interventions['Latitude Permanence'] = cardiac_interventions['Latitude Permanence'].apply(lambda x: adjust_coordinate(x, latitude_range))
    cardiac_interventions['Latitude Intervention'] = cardiac_interventions['Latitude Intervention'].apply(lambda x: adjust_coordinate(x, latitude_range))

# Apply the adjustment function to Longitude Permanence and Longitude Intervention
while any((cardiac_interventions['Longitude Permanence'] < longitude_range[0]) | (cardiac_interventions['Longitude Permanence'] > longitude_range[1])):
    cardiac_interventions['Longitude Permanence'] = cardiac_interventions['Longitude Permanence'].apply(lambda x: adjust_coordinate(x, longitude_range))
    cardiac_interventions['Longitude Intervention'] = cardiac_interventions['Longitude Intervention'].apply(lambda x: adjust_coordinate(x, longitude_range))
```

Adding one column to the table in which the time between T0 and T4 is computed (in minutes). The choice fell on T0-T4 because that's the time within the patient can potentially be stabilized.

```python
cardiac_interventions['T3_T0_difference_minutes'] = (cardiac_interventions['T3'] - cardiac_interventions['T0']).dt.total_seconds() / 60
```

Cleaning the dataset: excluding
- Cases in which there is an error in tracking the time ( 0 min > time > 180 min), or in which the time difference between T0 and T3 is not present
- Cases in which the patient was abandoned because of 'Error', 'Geannuleerd', 'Kwaadwillig', 'Vervoerd door derden', 'Vervoerd door politie', 'Zonder patient'

```python
condition1 = (cardiac_interventions['T3_T0_difference_minutes'] >= 0) & (cardiac_interventions['T3_T0_difference_minutes'] <= 180)
condition2 = ~cardiac_interventions['Abandon Reason'].isin(['Error', 'Geannuleerd', 'Kwaadwillig', 'Vervoerd door derden', 'Vervoerd door politie', 'Zonder patient'])

filtered_cardiac_interventions = cardiac_interventions[condition1 & condition2]

filtered_cardiac_interventions.head(5)
```

Now we need to filter out the columns which have the same code; we will:
- Add a new column called 'Vehicle Type 2' which will store the eventual second vehicle type arriving to the patient
- Check if either of the rows stored a death of the patient, and if it's the case it stores the resulting line as dead

```python
# Ensure the DataFrame is sorted by 'Mission ID' and index
filtered_cardiac_interventions = filtered_cardiac_interventions.sort_values(by=['Mission ID']).reset_index(drop=True)

# Process the DataFrame
def process_cardiac_interventions(df):
    # Create an empty 'Vehicle Type 2' column
    df['Vector Type 2'] = None

    # Iterate through the DataFrame
    rows_to_delete = []
    for i in range(len(df) - 1):
        if df.loc[i, 'Mission ID'] == df.loc[i + 1, 'Mission ID']:
            # Copy 'Vehicle Type' from the second row to 'Vehicle Type 2' in the first row
            df.loc[i, 'Vector Type 2'] = df.loc[i + 1, 'Vector Type']
            
            # Update 'Abandon Reason' if one of the rows has 'Overleden'
            if df.loc[i, 'Abandon Reason'] == 'Overleden' or df.loc[i + 1, 'Abandon Reason'] == 'Overleden':
                df.loc[i, 'Abandon Reason'] = 'Overleden'
            
            # Mark the second row for deletion
            rows_to_delete.append(i + 1)
    
    # Delete the marked rows
    df = df.drop(rows_to_delete).reset_index(drop=True)
    
    return df

# Apply the processing function
filtered_cardiac_interventions = process_cardiac_interventions(filtered_cardiac_interventions)
```

Added one column to show if the patient is dead or not

```python
filtered_cardiac_interventions['Dead'] = np.where(filtered_cardiac_interventions['Abandon Reason'] =='Overleden', 1, 0)
```

Now, this new version of the DF is saved and stored.
